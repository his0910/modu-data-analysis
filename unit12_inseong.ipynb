```python
import csv
f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)
result = []

for row in data:
    row[4:] = map(int, row[4:])
    result.append(row[10])

print(len(result))
print(result)
```

```python
# 이 데이터를 막대그래프로 표현하기
import matplotlib.pyplot as plt
plt.bar(range(len(result)), result)
plt.show()
```

```python
# 최다 인원 역 찾기
import csv
f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)
mx = 0
mx_station = ''

for row in data:
    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > mx :
        mx = sum(row[10:15:2])
        mx_station = row[3] + '('+row[1]+')'

print(mx_station)
```

```python
# 출근 시간대 사람들이 가장 많이 타고 내리는 역 찾기

import csv
f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)
mx = 0
mx_station = ''

for row in data:
    row[4:] = map(int, row[4:])
    a = row[11:16:2]
    
    if sum(a) > mx :
        mx = sum(a)
        mx_station = row[3] + '('+row[1]+')'

print(mx_station, mx)
```

```python
# 밤 11시에 사람들이 가장 많이 타는 역은 어디일까?

import csv
f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)
mx = 0
mx_station = ''
t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하세요? :'))

for row in data:
    row[4:] = map(int, row[4:])
    a = row[4+(t-4)*2] # 입력 받은 시각의 승차 인원 값 추출하기
    
    if a > mx :
        mx = a
        mx_station = row[3] + '('+row[1]+')'

print(mx_station, mx)
```

```python
# 시간대별로 사람들이 가장 많이 타고 내리는 역은 어디일까

import csv
f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)
mx = [0] * 24 # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station = [''] * 24 # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[j*2 + 4] # j와 인덱스 번호 사이의 관계식 활용
        
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3]
            
print(mx_station)
print(mx)
```

```python
# 이 데이터를 바탕으로 막대그래프 그리기

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
# 시간대별로 하차 인원이 가장 많은 역을 찾는 코드

import csv
import matplotlib.pyplot as plt

f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)
mx = [0] * 24 # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station = [''] * 24 # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        b = row[5 + j * 2] # j와 인덱스 번호 사이의 관계식 활용
        
        if b > mx[j] :
            mx[j] = b
            mx_station[j] = row[3] + '( ' +str(j+4)+ ' )'

plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(24), mx, color = 'b') # 막대그래프 속성 변경
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
# 모든 지하철역에서 시간대별 승하차 인원을 모두 더하면
# 지하철 시간대별 승하차 인원 추이를 나타내는 코드

import csv
import matplotlib.pyplot as plt

f = open('subwaytime_202111.csv')
data = csv.reader(f)

next(data)
next(data)

s_in = [0] * 24 # 승차 인원 저장 리스트 초기화
s_out = [0] * 24 # 하차 인원 저장 리스트 초기화

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4 + i*2]
        s_out[i] += row[5 + i*2]

plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')

plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')

plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
```
