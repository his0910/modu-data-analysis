```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(-int(i)) # 남성 데이터를 리스트 m에 저장
        for i in row[106:]:
            f.append(int(i)) # 여성 데이터를 리스트 f에 저장

# 데이터 시각화하기
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 위 그래프의 아래에서 -400, -200 부분이 깨져있음. 그래서 이렇게 수정함
plt.title(name+' 지역의 남녀 성별 인구 분포')
       
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()

print(len(m), len(f))
```

```python
# 위의 오류는? 지정된 x축의 개수(101개)와 찾은 데이터 개수가 일치하지 않아서 발생한 오류
# 나는 10행에서 'name' 을 name 으로 바꾸고 실행했더니 오류 없이 잘 되었음(애초에 잘못 입력한거였어)
```

```python
# 비율 및 범례 표시하기
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic') # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233] # 데이터
label = ['A형', 'B형', 'AB형', 'O형'] # 레이블

plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%') # autopct 속성
plt.legend()
plt.show()

# autopct : auto percent의 줄임말, 어떤 형태로 값을 표시할지 작성하면 각 항목의 비율을 자동으로 계산해서 표시
# 소수점 아래 둘째 자리에서 반올림한 값을 표시하려면 : %.1f%%
```

```python
# 색 및 돌출 효과 정하기
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic') # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233] # 데이터
label = ['A형', 'B형', 'AB형', 'O형'] # 레이블
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']

plt.axis('equal')
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0)) # autopct 속성
plt.legend()
plt.show()
```

```python
# 세곡동의 성별 인구 비율 표현하기
import csv
f = open('gender.csv')
data = csv.reader(f)

size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3]) # 3~103번까지 데이터 더하기
            f += int(row[i+106]) # 106~206번까지 데이터 더하기
        break

size.append(m)
size.append(f)
print(size)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic') # 그래프에 한글 표시
color = ['crimson', 'darkcyan']

plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)

plt.title(name+' 지역의 남녀 성별 비율')
plt.show()
```

```python
# 전체 코드
import csv
f = open('gender.csv')
data = csv.reader(f)

size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if name in row[0]:
        m = 0 # 남성 인구수를 누적해서 더할 변수 초기화하기
        f = 0 # 여성 인구수를 누적해서 더할 변수 초기화하기
        for i in range(101):
            m += int(row[i+3]) # 남성 인구수를 누적해서 더하기
            f += int(row[i+106]) # 여성 인구수를 누적해서 더하기
        break # 반복 종료

size.append(m) # 남성 인구수를 size 리스트에 더하기
size.append(f) # 여성 인구수를 size 리스트에 더하기

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
color = ['crimson', 'darkcyan'] # 색상 설정하기

plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)

plt.title(name+' 지역의 남녀 성별 비율') # 제목 설정하기
plt.show()
```
