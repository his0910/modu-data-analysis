```python
# UNIT 08 인구 구조를 다양한 형태로 시각화하기
# 2021.12.15
```

```python
import csv

f = open('age_gangnam.csv')
data = csv.reader(f)
result = [] # 빈 리스트 만들기
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data :
    if 'name' in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',',''))) # 해당 연령의 인구수 리스트에 순서대로 저장하기

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name+' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
# barh 함수 : 막대그래프를 수직이 아닌 수평 방향으로 그리는 함수
import csv
f = open('age_gangnam.csv')
data = csv.reader(f)

result = []
for row in data :
    if '세곡' in row[0]:
        for i in row[3:]:
            result.append(int(i))
            
import matplotlib.pyplot as plt
plt.barh(range(101), result)
plt.show()
```

```python
# 항아리 모양의 그래프 그리기
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '세곡' in row[0]:
        for i in range(0, 101):
            m.append(int(row[i+3]))
            f.append(int(row[-(i+1)]))

f.reverse() # reverse() 함수는 리스트의 값을 역순으로 재배열하는 함수
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '세곡' in row[0]:
        for i in row[3:104]:
            m.append(int(i)) # 남성 데이터를 리스트 m에 저장
        for i in row[106:]:
            f.append(int(i)) # 여성 데이터를 리스트 f에 저장

# 데이터 시각화하기
import matplotlib.pyplot as plt
plt.barh(range(101), m)
plt.barh(range(101), f)
plt.show()
```

```python
# 남성 데이터를 음수로 바꾸어서 남자 왼쪽 / 여자 오른쪽에 보기 좋게 시각화
# 그리고 그래프에 간단하게 제목과 범례를 추가

import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '세곡' in row[0]:
        for i in row[3:104]:
            m.append(-int(i)) # 남성 데이터를 리스트 m에 저장
        for i in row[106:]:
            f.append(int(i)) # 여성 데이터를 리스트 f에 저장

# 데이터 시각화하기
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.title('세곡동 지역의 남녀 성별 인구 분포')
       
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```

```python
# 우리 동네 인구 구조를 항아리 모양 그래프로 그리기

import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data:
    if 'name' in row[0]:
        for i in row[3:104]:
            m.append(-int(i)) # 남성 데이터를 리스트 m에 저장
        for i in row[106:]:
            f.append(int(i)) # 여성 데이터를 리스트 f에 저장

# 데이터 시각화하기
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 위 그래프의 아래에서 -400, -200 부분이 깨져있음. 그래서 이렇게 수정함
plt.title(name+' 지역의 남녀 성별 인구 분포')
       
plt.barh(range(101), m, label = '남성')
plt.barh(range(101), f, label = '여성')
plt.legend()
plt.show()
```
