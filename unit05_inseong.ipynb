```python
# 데이터 리스트에 저장하기
import csv
f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
    if row[-1] != '': # 최고 기온 데이터 값이 존재한다면(= 공백''이 아니면)
        result.append(float(row[-1])) # result 리스트에 실수 형태의 최고 기온 값을 추가
print(result)
```

```python
# 데이터 시각화 하기
# 1. matplotlib 라이브러리를 임포트하고
# 2. plot() 함수에 데이터를 넣고
# 3. show() 함수로 보여주면 완성
```

```python
import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '':
        result.append(float(row[-1]))

plt.plot(result, 'r') # result 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] == '08': # 8월에 해당하는 값이라면 
            result.append(float(row[-1]))

plt.plot(result, 'hotpink') # result 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '10' :
            result.append(float(row[-1]))

plt.plot(result, 'hotpink') # result 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '':
        if 1983 <= int(row[0].split('-')[0]): # 1983년 이후 데이터라면
            if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '10' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.plot(high, 'hotpink')
plt.plot(low, 'skyblue')

plt.rc('font', family = 'Malgun Gothic')
plt.title('my birthday')
plt.show()
```
