```python
import matplotlib.pyplot as plt
# as : alias(별명)
```

```python
# plot() : 직선, 꺾은선 형태의 그래프를 그릴 때 사용하는 함수
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

```python
# 그래프에 옵션 추가하기
plt.title('plotting')
plt.plot([10, 20, 30, 40])
plt.show()

# 제목 한글 설정 방법은 나중에
```

```python
# 그래프에 범례(legend) 넣기
plt.title('plotting')
plt.plot([10, 20, 30, 40], label = 'asc') # 증가를 의미하는 asc
plt.plot([40, 30, 20, 10], label = 'desc') # 감소를 의미하는 desc

plt.legend()
plt.show()
```

```python
# 범례 위치를 지정하려면?
# plt.legend(loc = 5) 이렇게 해라
```

```python
# 그래프 색상 바꾸기
plt.title('color') # 제목 설정

plt.plot([10, 20, 30, 40], color = 'skyblue', label = 'skyblue')
plt.plot([40, 30, 20, 10], color = 'pink', label = 'pink')

plt.legend(loc = 8)
plt.show()
```

```python
# 그래프 선 모양 바꾸기
plt.title('linestyle')

plt.plot([10, 20, 30, 40], color = 'r', linestyle = '--', label = 'dashed')
plt.plot([40, 30, 20, 10], color = 'g', linestyle = ':', label = 'dotted')

plt.legend(loc = 3)
plt.show()

# 색과 선 모양을 동시에 적는 형태로 코드를 작성할 수도 있다
# plt.plot([1,2,3,4], 'r--')
```

```python
# 마커 모양 바꾸기
plt.title('marker')
plt.plot([10, 20, 30 ,40], 'r.', label = 'circle') # 빨간색 원형 마커 그래프
plt.plot([40, 30, 20, 10], 'g^', label = 'triangle up') # 초록색 삼각형 마커 그래프

plt.legend()
plt.show()
```
