```python
# 유임 승차 비율이 가장 높은 역은 어디일까
# rate(비율) = 유임 승차인원 / 무임 승차인원

import csv
f = open('subwayfee_202111.csv')
data = csv.reader(f)
next(data)

mx = 0
rate = 0

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])
    rate = row[4] / row[6]
    if rate > mx :
        mx = rate
print(mx)
```

```python
# 0으로 나눴을 때 발생하는 오류 수정
import csv
f = open('subwayfee_202111.csv')
data = csv.reader(f)
next(data)

mx = 0
rate = 0

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])
        
    if row[6] != 0:
        rate = row[4] / row[6]
        if rate > mx :
            mx = rate
            print(row, round(rate, 2))
```

```python
# 유임 승차 비율 계산식 변경
# (변경후) rate = 유임 승차 인원 / 전체 인원
import csv
f = open('subwayfee_202111.csv')
data = csv.reader(f)
next(data)

mx = 0
rate = 0

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])
        
    if row[6] != 0 and (row[4] + row[6]) > 100000 :
        rate = row[4] / (row[4] + row[6])
        if rate > mx :
            mx = rate
            print(row, round(rate, 2))
        
print(mx)
```

```python
# 유임 승차 비율이 가장 높은 역 찾기
import csv

f = open('subwayfee_202111.csv')
data = csv.reader(f)
next(data)

mx = 0
rate = 0
mx_station = ''

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])
    
    if row[6] != 0 and (row[4] + row[6]) > 100000 :
        rate = row[4] / (row[4] + row[6])
        if rate > mx :
            mx = rate
            mx_station = row[3] + ' ' + row[1]
            
print(mx_station, round(mx*100, 2))
```

```python
# 유무임 승하차 인원이 가장 많은 역은 어디일까
import csv

f = open('subwayfee_202111.csv')
data = csv.reader(f)
next(data)

mx = [0] * 4
mx_station = [''] * 4
label = ['유임승차', '유임하차', '무임승차', '무임하차']

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])
        if row[i] > mx[i-4]:
            mx[i-4] = row[i]
            mx_station[i-4] = row[3] + ' ' + row[1]

for i in range(4):
    print(label[i] + ' : ' + mx_station[i], mx[i])
```

```python
# 모든 역의 유무임 승하차 비율은 어떻게 될까
import csv
import matplotlib.pyplot as plt

f = open('subwayfee_202111.csv')
data = csv.reader(f)
next(data)
label = ['유임승차', '유임하차', '무임승차', '무임하차']
c = ['#14ccc0', '#389993', '#FF1C6A', '#CC14AF']
plt.rc('font', family='Malgun Gothic')

for row in data:
    for i in range(4, 8):
        row[i] = int(row[i])

plt.figure(dpi=300)
plt.title(row[3]+' '+row[1])
plt.pie(row[4:8], labels=labe;, colors=c, autopct='%1.f%%')
plt.axis('equal')
plt.savefi
```
