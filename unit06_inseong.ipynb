```python
# 기온 데이터를 히스토그램으로 표현하기

import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
        
plt.hist(result, bins=100, color='r') # 히스토그램으로 나타내기
plt.show()
```

```python
# 8월 데이터만 뽑아서 히스토그램으로 나타내기

import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)
aug = [] # 8월의 최고 기온 값을 저장할 aug 리스트 생성

for row in data:
    month = row[0].split('-')[1] # -로 구분된 값 중 2번째 값을 month에 저장
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1])) # aug 리스트에 최고 기온 값을 추가
        
plt.hist(aug, bins=100, color='r') # 히스토그램으로 나타내기
plt.show()
```

```python
# 1월과 8월의 데이터를 히스토그램으로 시각화하기

import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    month = row[0].split('-')[1] # -로 구분된 값 중 2번째 값을 month에 저장
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1])) # aug 리스트에 최고 기온 값을 추가
        if month == '01':
            jan.append(float(row[-1]))
        
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```

```python
# 8월 최고 기온 데이터와 1월 최고 기온 데이터를 원소로 하는 리스트를 boxplot() 함수로 표현하기

import csv

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    month = row[0].split('-')[1] # -로 구분된 값 중 2번째 값을 month에 저장
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1])) # aug 리스트에 최고 기온 값을 추가
        if month == '01':
            jan.append(float(row[-1]))
            
import matplotlib.pyplot as plt
plt.boxplot([aug, jan])
plt.show()
```

```python
# 최고 기온 데이터를 월별로 구분하여 표현하기

import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)

# 월별 데이터를 저장할 리스트 month 12개 생성
month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
    if row[-1] != '':
        # 월과 같은 번호의 인덱스에 월별 데이터를 저장(예 : 1월 -> nonth[0])
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
# 8월 일별 기온 데이터를 상자 그림으로 표현하기

import csv
import matplotlib.pyplot as plt

f = open('temp_seoul.csv')
data = csv.reader(f)
next(data)

day = [] # 일별 데이터를 저장할 리스트 day 생성
for i in range(31):
    day.append([]) # day 리스트 내 31개의 리스트를 생성
    
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            day[int(row[0].split('-')[2])-1].append(float(row[-1])) # 최고 기온 값을 저장
            
plt.style.use('ggplot') # 그래프 스타일 지정
plt.figure(figsize=(10, 5), dpi=300) # 그래프 크기 수정
plt.boxplot(day, showfliers=False) # 아웃라이어 값은 생략

plt.show()
```
